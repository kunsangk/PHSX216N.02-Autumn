```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# === PART I: Converging Lens Data ===
d_o1 = np.array([20, 30, 40, 50, 60])
delta_d_o1 = np.array([0.141421356] * 5)

d_i1 = np.array([61.1, 29.6, 23.6, 20.9, 19.7])
delta_d_i1 = np.array([0.141421356] * 5)

product1 = d_o1 * d_i1
delta_product1 = np.abs(product1) * np.sqrt((delta_d_o1 / d_o1)**2 + (delta_d_i1 / d_i1)**2)
sum_di_do1 = d_o1 + d_i1

# Regression for converging lens
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(sum_di_do1, product1)

# === PART II: Diverging Lens Data ===
d_o2 = np.array([-12.7, -10.7, -8.7, -6.7, -4.7])
delta_d_o2 = np.array([0.141421356] * 5)

d_i2 = np.array([36.3, 23.9, 15.8, 10.1, 6.3])
delta_d_i2 = np.array([0.141421356] * 5)

product2 = d_o2 * d_i2
delta_product2 = np.abs(product2) * np.sqrt((delta_d_o2 / np.abs(d_o2))**2 + (delta_d_i2 / d_i2)**2)
sum_di_do2 = d_o2 + d_i2

# Regression for diverging lens
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(sum_di_do2, product2)

# === Plot: Converging Lens ===
plt.figure(figsize=(8, 5))
plt.errorbar(sum_di_do1, product1, yerr=delta_product1, fmt='o', label='Measured Data')
plt.plot(sum_di_do1, slope1 * sum_di_do1 + intercept1, label=f'Fit: f = {slope1:.2f}  {std_err1:.2f} cm')
plt.xlabel(r'$d_o + d_i$ (cm)')
plt.ylabel(r'$d_o \cdot d_i$ (cm)')
plt.title('Converging Lens: Focal Length Determination')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Plot: Diverging Lens ===
plt.figure(figsize=(8, 5))
plt.errorbar(sum_di_do2, product2, yerr=delta_product2, fmt='o', label='Measured Data')
plt.plot(sum_di_do2, slope2 * sum_di_do2 + intercept2, label=f'Fit: f = {slope2:.2f}  {std_err2:.2f} cm')
plt.xlabel(r'$d_o + d_i$ (cm)')
plt.ylabel(r'$d_o \cdot d_i$ (cm)')
plt.title('Diverging Lens: Focal Length Determination')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# === Part I: Force vs. Displacement (Corrected for mg units) ===

# Displacement (z in meters)
delta_z_part1 = np.array([0.008, 0.015, 0.024, 0.031, 0.038])

mass_g = np.array([1, 2, 3, 4, 5])         # originally in grams
mass_kg = mass_g * 1e-3                    # convert to kg
g = 9.81                                   # m/s^2
forces_corrected = mass_kg * g             # N
force_uncertainty = np.array([0.00098] * 5)  # N (0.1g converted)

# Linear regression
slope_k, intercept_k, _, _, std_err_k = linregress(delta_z_part1, forces_corrected)

# Plot Part I
plt.figure()
plt.errorbar(delta_z_part1, forces_corrected, yerr=force_uncertainty, fmt='o', capsize=5, label='Data with Uncertainty')
plt.plot(delta_z_part1, slope_k * delta_z_part1 + intercept_k,
         label=f'Fit: F = {slope_k:.2f}z + {intercept_k:.2f}', color='purple')
plt.xlabel('Displacement z (m)')
plt.ylabel('Force F (N)')
plt.title('Corrected Part I: Force vs. Displacement (mg units)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Part III: Displacement vs. Current (with negative z) ===

# Current (A) and displacement (z in m)
current = np.array([0.201, 0.406, 0.601, 0.803, 1.006, 1.203, 1.4])
delta_z = np.array([-0.002, -0.004, -0.007, -0.009, -0.012, -0.013, -0.014])
delta_z_uncertainty = np.array([0.0001] * 7)

# Linear regression
slope_mu, intercept_mu, _, _, std_err_mu = linregress(current, delta_z)

# Plot Part III
plt.figure()
plt.errorbar(current, delta_z, yerr=delta_z_uncertainty, fmt='o', capsize=5, label='Data with Uncertainty')
plt.plot(current, slope_mu * current + intercept_mu,
         label=f'Fit: z = {slope_mu:.5f}I + {intercept_mu:.5f}', color='red')
plt.xlabel('Current I (A)')
plt.ylabel('Displacement z (m)')
plt.title('Updated Part III: Displacement vs. Current (Anti-parallel Coils)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
